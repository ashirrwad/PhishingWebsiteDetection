### Importing some packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Loading the data

```python
df = pd.read_csv("C:\\Users\\Vahe\\PycharmProjects\\pythonProject\\Training_Batch_Files\\phishing_09032022_ 082200.csv")
df.head()
```

### EDA

Checking for general info

```python
df.info()
```

Checking missing values

```python
df.isnull().sum().sum()
```

Checking for zero variance features

```python
zero_variance_features = []
for col in df.columns:
    if df[col].std() == 0:
        zero_variance_features.append(col)

print(zero_variance_features)
```

#### Univariate analysis

Creating some necessary lists

```python
# Creating list of categorical variables including the target (phishing)
cat_list = ['email_in_url', 'domain_in_ip', 'server_client_domain', 
            'tld_present_params', 'domain_spf', 'tls_ssl_certificate',
            'url_google_index', 'domain_google_index', 'url_shortened',
            'phishing']

# Creating features' list excluding zero variance features
feat_list = [col for col in df.columns if col not in zero_variance_features]
```

```python
fig = plt.figure(figsize=(12, 8))
i = 1
for col in feat_list[:16]:
    plt.subplot(4, 4, i)
    if col in cat_list:
        sns.countplot(x=col, data=df)
    else:
        sns.kdeplot(x=col, data=df)
        
    i += 1
    
plt.tight_layout()
```
